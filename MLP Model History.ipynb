```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
#Read Model History

model_hist = np.load('./Model_history/mlp_hist.npy',allow_pickle='TRUE').item()
print(model_hist.keys())
```

```python
epochs=[1,2,3,4,5]
```

```python
plt.plot(epochs,model_hist['loss'],label='Loss')
plt.scatter(epochs,model_hist['loss'])
plt.plot(epochs,model_hist['val_loss'],label='Validation Loss')
plt.scatter(epochs,model_hist['val_loss'])
plt.xlabel('Epochs',fontsize=14)
plt.ylabel('Loss',fontsize=14)
plt.title('Epoch Vs Loss',fontsize=20)
plt.legend()
plt.show()
```

```python
plt.plot(epochs,model_hist['accuracy'],label='Accuracy')
plt.scatter(epochs,model_hist['accuracy'])
plt.plot(epochs,model_hist['val_accuracy'],label='Validation Accuracy')
plt.scatter(epochs,model_hist['val_accuracy'])
plt.xlabel('Epochs',fontsize=14)
plt.ylabel('Accuracy',fontsize=14)
plt.title('Epoch Vs Accuracy',fontsize=20)
plt.legend()
plt.show()
```
